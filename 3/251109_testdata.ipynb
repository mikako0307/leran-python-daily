```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
import numpy as np
```

```python
# csvファイルをデータフレームとして読み込む
df = pd.read_csv("/content/593771db-c252-4a0d-8b5b-6b7b055a192c.csv")

df.head()
```

```python
# value列を抽出しnumpy配列に変換
value_np = df['Value'].to_numpy()

# 最小値と最大値を取得
min_val = np.min(value_np)
max_val = np.max(value_np)

normalized_value = (value_np - min_val) / (max_val - min_val)

# 結果をDataFrameに追加
df['Normalized_value'] = normalized_value

```

```python
# 正規化後のデータをplt.plotを用いてグラフ化する
plt.plot(df['Time'], normalized_value, label='Normalized Value')
plt.xlabel('Time')
plt.ylabel('Normalized Value')
plt.title('正規化されたValueの推移')
plt.show
```
