```python
import pandas as pd
# DataFrame に変換
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
# 日本語にする
rename_dict = {
    "Elevation": "標高",
    "Aspect": "斜面方位",
    "Slope": "傾斜角度",
    "Horizontal_Distance_To_Hydrology": "水系までの水平距離",
    "Vertical_Distance_To_Hydrology": "水系までの垂直距離",
    "Horizontal_Distance_To_Roadways": "道路までの水平距離",
    "Horizontal_Distance_To_Fire_Points": "火災地点までの水平距離",
    "Hillshade_9am": "午前9時の日照指数",
    "Hillshade_Noon": "正午の日照指数",
    "Hillshade_3pm": "午後3時の日照指数",
    "Wilderness_Area_0": "荒野区分0",
    "Wilderness_Area_1": "荒野区分1",
    "Wilderness_Area_2": "荒野区分2",
    "Wilderness_Area_3": "荒野区分3",
}
# Soil_Type の40種類をまとめて辞書に追加
for i in range(40):
    rename_dict[f"Soil_Type_{i}"] = f"土壌区分{i}"


```

```python
# カラム名を日本語に変換
df = df.rename(columns=rename_dict)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd


# 特徴量と目的変数に分ける
X = df.drop(columns=['森林被覆タイプ']).to_numpy()
y = df['森林被覆タイプ'].to_numpy()

# スケーリング/標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# ランダムフォレスト
from sklearn.model_selection import train_test_split

# 分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

# モデル作成
rf = RandomForestClassifier(
    n_estimators=100,   # 木の数
    max_depth=None,     # 深さ制限なし
    random_state=42,
    n_jobs=-1           # CPU並列処理
)

# 学習
rf.fit(X_train, y_train)

# 予測
y_pred = rf.predict(X_test)
```

```python
# 評価指標
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_test)
```

```python
# 重要度の確認
import matplotlib.pyplot as plt
import numpy as np

# 学習済みモデル rf から重要度を取得
importances = rf.feature_importances_

# 特徴量名（dfの列名から目的変数を除いたもの）
feature_names = df.drop(columns=['森林被覆タイプ']).columns

# 重要度をソート
indices = np.argsort(importances)[::-1]

# 上位20特徴量を棒グラフで可視化
plt.figure(figsize=(10,6))
plt.title("Feature Importances (Top 20)")
plt.bar(range(20), importances[indices[:20]], align="center")
plt.xticks(range(20), feature_names[indices[:20]], rotation=90)
plt.tight_layout()
plt.show()
```

特徴量が多い、クラス分類が多い、データ数が多い→ランダムフォレストを使用
標高が森林被覆タイプへの影響が最もあることが分かった。
道路までの水平距離、火災地点までの水平距離、水系までの水平距離・垂直距離、日照指数が影響を与えていることが分かった。
