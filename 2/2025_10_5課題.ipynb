```python
dataset_a = [10, 25, 40, 50, 75]
dataset_b = [12, 18, 24, 30, 36, 42, 56, 36, 82, 4]

# --- 以下に記述 -----------------------
class DataAnalizer:
    # コンストラクタでデータセットの属性(属性名はdataset)を定義してください。
    def __init__(self, dataset):
        self.dataset = dataset

    # データセットの要素数を戻り値で返すメソッド(メソッド名はLength)を作成してください。
    def Length(self):
        return len(self.dataset)

    # 全要素の合計値を戻り値で返すメソッド(メソッド名はSum)を作成してください。
    def Sum(self):
        return sum(self.dataset)

    # 【発展課題】全要素の平均値を戻り値で返すメソッド(メソッド名はAverage)を作成してください。
    # ヒント：上記で定義したメソッドを使えます。
    def Average(self):
        return self.Sum() / self.Length()


# -------------------------------------------

a = DataAnalizer(dataset_a)
b = DataAnalizer(dataset_b)

data_list = [a, b]
for data in data_list:
    print("-" * 30)
    print(f"データセット: {data.dataset}")
    print(f"要素数: {data.Length()}")
    print(f"合計: {data.Sum()}")
    print(f"平均: {data.Average()}")
```
